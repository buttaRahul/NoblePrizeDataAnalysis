```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
nobel = pd.read_csv("C:\\Users\\butta\\Desktop\\project\\nobel visualization\\nobel.csv")

nobel.head(6)
```

```python
display(len(nobel.organization_country))

display(nobel.sex.value_counts())

nobel.birth_country.value_counts()[:10]
```

```python
nobel['usa_born_winner'] = nobel.birth_country == 'United States of America'
nobel['decade'] = (np.floor(nobel.year/10) * 10).astype('int64')
prop_usa_winners = nobel.groupby('decade', as_index=False)['usa_born_winner'].mean()

prop_usa_winners
```

```python
# Setting the plotting theme
sns.set()
# and setting the size of all plots.
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [11, 7]

# Plotting USA born winners 
ax = sns.lineplot(x='decade', y='usa_born_winner', data=nobel)

# Adding %-formatting to the y-axis
from matplotlib.ticker import PercentFormatter
ax.yaxis.set_major_formatter(PercentFormatter())
```

```python
nobel['female_winner'] = nobel["sex"]=="Female"
prop_female_winners = nobel.groupby(["decade","category"], as_index=False)["female_winner"].mean()

ax = sns.barplot(x="decade", y="female_winner", hue="category", data=prop_female_winners)

plt.title("The proportion of female winners per decade by categories", 
          fontdict={'fontsize': 14},
          fontweight ='bold',
          loc='center')
plt.ylabel("Female winners ratio")
plt.xlabel("Decade")
plt.legend(loc='upper left')

from matplotlib.ticker import PercentFormatter
ax.yaxis.set_major_formatter(PercentFormatter(1.0))
```

```python
nobel['female_winner'] = nobel.sex == 'Female'
prop_female_winners = nobel.groupby(['decade', 'category'], as_index=False)['female_winner'].mean()

ax = plt.subplot()
ax = sns.lineplot(x='decade', y='female_winner', data=prop_female_winners, hue='category')

ax.yaxis.set_major_formatter(PercentFormatter())
```
